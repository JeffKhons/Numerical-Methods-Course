```python
import numpy as np
import matplotlib.pyplot as plt

def function(x):
    f = -1*(x**2) + 6*x + 56
    return f

a = -2  # Lower limit of integration
b = 11   # Upper limit of integration
n = 81  # Number of subintervals

# Calculate the step size
h = (b - a) / n

# Create an array of evenly spaced grid points
x = np.linspace(a, b, n+1)

# Calculate the function values at the grid points
f = function(x)

# Calculate the trapezoid rule approximation
Integral_trapezoid = h * (np.sum(f) - 0.5 * (f[0] + f[-1]))

# Calculate the error
error = Integral_trapezoid - 632.666

print('Integral_trapezoid:', Integral_trapezoid)
print('error:', error)

# Plot the function and the trapezoid integral procedure
plt.plot(x, function(x), linewidth=2, c='r')

for i in range(0, n-1):
    c = ['darkorange', 'forestgreen', 'royalblue']
    x_range = [x[i], x[i], x[i+1], x[i+1]]
    y_range = [0, f[i], f[i+1], 0]
    plt.fill(x_range, y_range, color=c[i%3], alpha=0.5)

plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Trapezoid Integral Procedure')
plt.grid(True)
plt.show()
```
