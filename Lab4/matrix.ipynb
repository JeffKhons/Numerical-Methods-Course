```python
import random
from copy import deepcopy

class Matrix:
    def __init__(self, nrows, ncols):
        """Construct a (nrows X ncols) matrix"""
        self.nrows = nrows
        self.ncols = ncols
        self.matrix = [[random.randint(0, 10) for _ in range(ncols)] for _ in range(nrows)]

    def add(self, m):
        """return a new Matrix object after summation"""
        if self.nrows != m.nrows or self.ncols != m.ncols:
            print("Unacceptable adding: Matrices are not of the same size")
            return None

        result = Matrix(self.nrows, self.ncols)
        result.matrix = [[self.matrix[i][j] + m.matrix[i][j] for j in range(self.ncols)] for i in range(self.nrows)]
        return result

    def sub(self, m):
        """return a new Matrix object after subtraction"""
        if self.nrows != m.nrows or self.ncols != m.ncols:
            print("Unacceptable subtraction: Matrices are not of the same size")
            return None

        result = Matrix(self.nrows, self.ncols)
        result.matrix = [[self.matrix[i][j] - m.matrix[i][j] for j in range(self.ncols)] for i in range(self.nrows)]
        return result

    def mul(self, m):
        """return a new Matrix object after multiplication"""
        if self.ncols != m.nrows:
            print("Unacceptable multiplication: The number of columns in the first matrix must be equal to the number of rows in the second matrix")
            return None

        result = Matrix(self.nrows, m.ncols)
        result.matrix = [[sum(self.matrix[i][k] * m.matrix[k][j] for k in range(self.ncols)) for j in range(m.ncols)] for i in range(self.nrows)]
        return result

    def transpose(self):
        """return a new Matrix object after transpose"""
        result = Matrix(self.ncols, self.nrows)
        result.matrix = [[self.matrix[j][i] for j in range(self.nrows)] for i in range(self.ncols)]
        return result

    def display(self):
        """Display the content in the matrix"""
        for row in self.matrix:
            print(row)


# Example usage with user input:
a_rows = int(input("Enter A matrix's rows: "))
a_cols = int(input("Enter A matrix's cols: "))
print("Matrix A({}, {}):".format(a_rows, a_cols))
A = Matrix(a_rows, a_cols)
A.display()

b_rows = int(input("Enter B matrix's rows: "))
b_cols = int(input("Enter B matrix's cols: "))
print("Matrix B({}, {}):".format(b_rows, b_cols))
B = Matrix(b_rows, b_cols)
B.display()

print("="*10, 'A + B', "="*10)
result = A.add(B)
if result is not None:
    result.display()

print("="*10, 'A - B', "="*10)
result = A.sub(B)
if result is not None:
    result.display()

print("="*10, 'A * B', "="*10)
result = A.mul(B)
if result is not None:
    result.display()
    
print("="*5, "the transpose of A*B", "="*5)
result = result.transpose()
if result is not None:
    result.display()
```
